```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing and displaying dataset
# uses this dataset -> https://www.kaggle.com/nehalbirla/vehicle-dataset-from-cardekho?select=Car+details+v3.csv
dataset = pd.read_csv('/content/drive/MyDrive/MSP/8th Task/Car details v3.csv')
dataset
```

```python
dataset.info()
```

```python
dataset['transmission'].unique()
```

```python
#Create Copy of Dataset
dataset_orig = dataset.copy()
# Importing required package
from sklearn.preprocessing import LabelEncoder
encode_x = LabelEncoder()
```

```python
dataset['fuel'] = encode_x.fit_transform(dataset['fuel'])
dataset['seller_type'] = encode_x.fit_transform(dataset['seller_type'])
dataset['owner'] = encode_x.fit_transform(dataset['owner'])
dataset['mileage'] = encode_x.fit_transform(dataset['mileage'])
dataset['engine'] = encode_x.fit_transform(dataset['engine'])
dataset['max_power'] = encode_x.fit_transform(dataset['max_power'])
dataset['torque'] = encode_x.fit_transform(dataset['torque'])
```

```python
# transmission is only either Manual or Automatic, so we can make it a dummy value
y = pd.get_dummies(dataset['transmission'], prefix='transmission')
y.head()
```

```python
dataset = pd.concat([dataset,pd.get_dummies(dataset['transmission'],prefix='transmission',drop_first=True)],axis=1)
dataset.drop(['transmission'],axis=1, inplace=True)
dataset.head()
```

```python
dataset
```

```python
# Converting transmission from object to integer
dataset['transmission_Manual'] = dataset['transmission_Manual'].astype(str).astype('int64')
```

```python
dataset['seats'] = dataset['seats'].fillna(dataset['seats'].median())
dataset['seats'] = dataset['seats'].astype(float).astype('int64')
```

```python
dataset.isna().any()
```

```python
dataset.info()
```

```python
features = ['year', 'km_driven', 'fuel', 'seller_type', 'owner', 'mileage', 'engine', 'max_power', 'torque', 'seats', 'transmission_Manual']
# features = ['km_driven']
target = 'selling_price'

X = dataset[features].values.reshape(-1, len(features))
Y = dataset[target].values
Y = Y.reshape(-1, 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_Y = StandardScaler()
X = sc_X.fit_transform(X)
Y = sc_Y.fit_transform(Y)
```

```python
# Spliting the datdset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test  = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.svm import SVR
regressor = SVR()
regressor.fit(X_train, Y_train)
```

```python
Y_pred = regressor.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import mean_squared_error
score = regressor.score(X, Y)
print("R-squared:", score)
print("MSE:", mean_squared_error(Y_test, Y_pred))
```
