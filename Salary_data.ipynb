```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
def plot_regression_line(scatAxis1, scatAxis2, plotAxis1, plotAxis2, scatColour, plotColour):
    """
    * Function draws the graph using the arguments given.

    scatAxis1, scatAxis2: the X and Y axis of the points
    plotAxis1, plotAxis2: the X and Y axis of the regression line
    scatColour, plotColour: the colours of the points and of the regression line

    returns: nothing
    """
    # plotting the actual points as scatter plot
    plt.scatter(scatAxis1, scatAxis2, color=scatColour)

    # plotting the regression line
    plt.plot(plotAxis1, plotAxis2, color=plotColour)

    # putting labels
    plt.xlabel("Years of Experience")
    plt.ylabel("Salary")

    # function to show plot
    plt.show()
```

```python
if __name__ == "__main__":
    # reading dataset
    data_sets = pd.read_csv('/content/drive/MyDrive/MSP/5th task/Salary_Data.csv')
    x = data_sets.iloc[:, :-1].values
    y = data_sets.iloc[:, 1].values

    # splitting dataset & displaying their shape
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1 / 3, random_state=0)
    print("Training data shapes:\n\tX:", x_train.shape, "| Y:", y_train.shape,
          "\nTesting data shapes:\n\tX:", x_test.shape, "| Y:", y_test.shape)

    # define & train regression model
    regressor = LinearRegression()
    regressor.fit(x_train, y_train)
    f_x = regressor.predict(x_train)

    # plotting training data
    plt.title("Salary vs Experience (Training data)")
    plot_regression_line(x_train, y_train, x_train, f_x, "green", "red")

    # calculating the intercept & slope of the best-fit line
    print('intercept:', round(regressor.intercept_, 3))
    print('slope:', round(regressor.coef_[0], 3))

    # calculating score for training data
    r_square = regressor.score(x_train, y_train)
    print('coefficient of determination (R^2) for Train Set:', round(r_square, 3))

    # plotting testing data
    plt.title("Salary vs Experience (Test data)")
    plot_regression_line(x_test, y_test, x_train, f_x, "blue", "red")

    # calculating score for testing data
    r_square_test = regressor.score(x_test, y_test)
    print('coefficient of determination (R^2) for Test Set:', round(r_square_test, 3))
```
